```r
library(tidyverse)
library(ggplot2)

astacus <- read_csv('/gstock/biolo_datasets/Others/Christelle/length_astacus.csv')
astacus$species <- "astacus"
marbled <- read_csv('/gstock/biolo_datasets/Others/Christelle/length_marbled.csv')
marbled$species <- "marbled"
concat_df <- rbind(astacus, marbled)

seq <- seq(400,3000, 100)
seq[length(seq)+1] = 50000
seq


df_astacus <- as.data.frame(table(cut(astacus$length, breaks = seq)))
df_astacus$species <- "astacus"

df_marbled <- as.data.frame(table(cut(marbled$length, breaks = seq)))
df_marbled$species <- "marbled"

df <- rbind(df_astacus, df_marbled)


ggplot(df) + geom_bar(mapping = aes(x=Var1, y=Freq, fill=species), stat='identity', position='dodge')


```
