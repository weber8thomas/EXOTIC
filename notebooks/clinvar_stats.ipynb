```python
import pandas as pd
```

```python
cv = pd.read_csv('/gstock/biolo_datasets/variation/variation_sets/clinvar/vcf_GRCh37/v2/clinvar_20210731.vcf.gz', compression='gzip', sep='\t', skiprows=27)
cv
```

```python
import numpy as np
cv['MC'] = cv['INFO'].apply(lambda r: [e.split('=')[1] for e in r.split(';') if 'MC' in e])
cv['MC'] = cv['MC'].apply(lambda r: r[0] if r else np.nan)
cv['CLNVC'] = cv['INFO'].apply(lambda r: [e.split('=')[1] for e in r.split(';') if 'CLNVC=' in e][0])
cv['CLNSIG'] = cv['INFO'].apply(lambda r: [e.split('=')[1] for e in r.split(';') if 'CLNSIG' in e][0])
cv
```

```python
raw_cv_series = cv.dropna(subset=['MC']).MC.apply(lambda r: [e.split('|')[1] for e in r.split(',') if 'SO' in e]).dropna().apply(lambda r : r[0] if r else np.nan).value_counts().sort_index()
raw_cv_series
```

```python
raw_cv_snv = cv.loc[cv['CLNVC'] == 'single_nucleotide_variant'].dropna(subset=['MC']).MC.apply(lambda r: [e.split('|')[1] for e in r.split(',') if 'SO' in e]).dropna().apply(lambda r : r[0] if r else np.nan).value_counts().sort_index().rename('SNV')
raw_cv_snv
```

```python
raw_cv_snv_path = cv.loc[(cv['CLNVC'] == 'single_nucleotide_variant') & (cv['CLNSIG'].str.contains('athogenic') == True)].dropna(subset=['MC']).MC.apply(lambda r: [e.split('|')[1] for e in r.split(',') if 'SO' in e]).dropna().apply(lambda r : r[0] if r else np.nan).value_counts().sort_index().rename('Pathogenic')
raw_cv_snv_path
```

```python
raw_cv_snv_path_conflict = cv.loc[(cv['CLNVC'] == 'single_nucleotide_variant') & (cv['CLNSIG'].str.contains('athogenic') == True) & (~cv['CLNSIG'].str.contains('onflicting') == True)].dropna(subset=['MC']).MC.apply(lambda r: [e.split('|')[1] for e in r.split(',') if 'SO' in e]).dropna().apply(lambda r : r[0] if r else np.nan).value_counts().sort_index().rename('Pathogenic_wt_conflicts')
raw_cv_snv_path_conflict
```

```python
raw_cv_snv_uncertain = cv.loc[(cv['CLNVC'] == 'single_nucleotide_variant') & (cv['CLNSIG'].str.contains('Uncertain') == True)].dropna(subset=['MC']).MC.apply(lambda r: [e.split('|')[1] for e in r.split(',') if 'SO' in e]).dropna().apply(lambda r : r[0] if r else np.nan).value_counts().sort_index().rename('Uncertain')
raw_cv_snv_uncertain
```

```python
concat = pd.concat([raw_cv_snv, raw_cv_snv_path, raw_cv_snv_path_conflict, raw_cv_snv_uncertain], axis=1).fillna(0).astype(int)
concat['Ratio_uncertain'] = 100 * (concat['Uncertain'] / concat['SNV'])
concat.loc["Sum"] = concat.sum(axis=0)

concat
```
