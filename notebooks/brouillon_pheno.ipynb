```python
import sweetviz as sv
import numpy as np
tmp_df = df.copy()
tmp_df['Disorder_classification'] = tmp_df['Disorder_classification'].apply(eval)
tmp_df['Pheno_Specific_Alt_details_Total_variants'] = tmp_df['Pheno_Specific_Alt_details_Total_variants'].astype(np.int64)
tmp_df = tmp_df.explode('Disorder_classification')
# feature_config = sv.FeatureConfig(force_num=['Pheno_Specific_Alt_details_Total_variants'])

tmp_df['Disorder_classification'] = tmp_df['Disorder_classification'].str.replace('Orphanet classification of rare ', '')
my_report = sv.analyze(tmp_df[['CCRS_Gene', 'Disorder_name', 'Disorder_type', 'Disorder_classification', 'Pheno_Specific_Const', 'Pheno_Specific_Const_nb', 
                               'Pheno_Specific_Alt', 'Pheno_Specific_Alt_nb', 'Pheno_Common', 'Pheno_Common_nb', 'Pheno_Specific_Const_details_Total_variants', 
                               'Pheno_Specific_Alt_details_Total_variants', 'Pheno_Common_details_Total_variants']])
my_report.show_html('test.html') # Default arguments will generate to "SWEETVIZ_REPORT.html"
```

```python
tmp_df['Disorder_classification'] = tmp_df['Disorder_classification'].str.strip()
tmp_df['Disorder_classification'] = tmp_df['Disorder_classification'].str.replace('Orphanet classification of ', '')
tmp_df = tmp_df.loc[tmp_df['Disorder_classification'] != 'genetic diseases']
import matplotlib.pyplot as plt
plt.style.use('ggplot')

plot = tmp_df.Disorder_classification.value_counts().plot.bar()
plot.set_ylabel('Nb of occurences')
plt.title('Orphanet classification disorder distribution')
```
