```python
def main_transform(x):
    x = x.dropna()

    def sigmoid(e):
        return 1 / (1 + math.exp(-e))

    vsigmoid = np.vectorize(sigmoid)

    def norm(e):
        return (2 * ((e - 0) / (1 - 0))) - 1

    vnorm = np.vectorize(norm)

    def remove_value(x, i):
        return [sub_e for j, sub_e in enumerate(x) if j != i]

    def modif_zscore(x):
        return [
            e_z * np.abs(e_x - np.median([e for i, e in enumerate(x) if i != j]))
            for (j, e_x), e_z in zip(enumerate(x), stats.zscore(x))
        ]

    z_score_x = modif_zscore(x)
    z_score_x = vsigmoid(z_score_x)
    z_score_x = vnorm(z_score_x)

    return z_score_x

def main_transform_without_weight(x):
    x = x.dropna()

    def sigmoid(e):
        return 1 / (1 + math.exp(-e))

    vsigmoid = np.vectorize(sigmoid)

    def norm(e):
        return (2 * ((e - 0) / (1 - 0))) - 1

    vnorm = np.vectorize(norm)

    def remove_value(x, i):
        return [sub_e for j, sub_e in enumerate(x) if j != i]

    def modif_zscore(x):
        return [
            e_z * np.abs(e_x - np.median([e for i, e in enumerate(x) if i != j]))
            for (j, e_x), e_z in zip(enumerate(x), stats.zscore(x))
        ]

    z_score_x = stats.zscore(x)
    z_score_x = vsigmoid(z_score_x)
    z_score_x = vnorm(z_score_x)

    return z_score_x
```

```python
from matplotlib.lines import Line2D

f, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

start, stop, nb = 0.48, 0.52, 52
random.seed(0)
case = [random.uniform(start, stop) for e in range(nb)]
std, mean = np.std(case), np.mean(case)
print(std, mean)
case = case + [0.4,0.6]
random.shuffle(case)
np.array(case)

ax[0][0].plot(case, color='blue')
ax[0][0].set_ylim(0,1)
ax[0][0].set_ylabel('pext')
ax[0][0].set_title('Case 0')
ax[1][0].plot(main_transform_without_weight(pd.Series(case)), color='grey')
ax[1][0].plot(main_transform(pd.Series(case)), color='red')
ax[1][0].axhline(-0.5, color='black', linestyle='--')
ax[1][0].axhline(0.5, color='black', linestyle='--')
ax[1][0].set_ylim(-1,1)
ax[1][0].legend(
    [
        Line2D([0], [0], color='grey', lw=1),
        Line2D([0], [0], color='red', lw=1)
    ], 
    [
        'dext (without weight)',
        'dext'
    ]
)

start, stop, nb = 0.18, 0.22, 50
random.seed(2)
case = [random.uniform(start, stop) for e in range(nb)]
std, mean = np.std(case), np.mean(case)
print(std, mean)
case = case + [0,0.6,0.4,1]
random.shuffle(case)
np.array(case)

ax[0][1].plot(case, color='blue')
ax[0][1].set_ylim(0,1)
ax[0][1].set_ylabel('pext')
ax[0][1].set_title('Case 1')
# ax[0][1].set_xticklabels([])
ax[1][1].plot(main_transform_without_weight(pd.Series(case)), color='grey')
ax[1][1].plot(main_transform(pd.Series(case)), color='red')
ax[1][1].axhline(-0.5, color='black', linestyle='--')
ax[1][1].axhline(0.5, color='black', linestyle='--')
ax[1][1].set_ylim(-1,1)
# handles, labels = ax[1][1].get_legend_handles_labels()
ax[1][1].legend(
    [
        Line2D([0], [0], color='grey', lw=1),
        Line2D([0], [0], color='red', lw=1)
    ], 
    [
        'dext (without weight)',
        'dext'
    ]
)
```
