```python
def stats_or(cat_const, cat_alt, others_const, others_alt):
    
    odd_ratio, pvalue = scipy.stats.fisher_exact([[cat_const, cat_alt], [others_const, others_alt]])
    se = np.sqrt((1/cat_const) + (1/others_const) + (1/cat_alt) + (1/others_alt))
    lower = np.exp(np.log(odd_ratio) - 1.96 *se)
    upper = np.exp(np.log(odd_ratio) + 1.96 *se)

    
    d = {
            'OR' : odd_ratio,
            'SE' : se,
            'IC_low' : lower,
            'IC_up' : upper,
            'pvalue' : pvalue,
        }
    return d
```
