```python
dict_mrna = collections.defaultdict(dict)
current_elem = str()
for j, row in tqdm(mrna_df.iterrows()):
    if row['Element'] == 'gene':
        gene = row['Name']
    elif row['Element'] == 'mRNA':
        current_elem = row['Name']
        dict_mrna[gene][current_elem] = list()
    elif row['Element'] == 'CDS':
        dict_mrna[gene][current_elem].append(row['ranges'])
```

```python
turn_mrna_dict = collections.defaultdict(list)
for i, (k, v) in tqdm(enumerate(dict_mrna.items())):
    for mrna, cds_list in v.items():
        for cds in cds_list:
            turn_mrna_dict[k + '_' + cds].append(mrna)
```

```python
def basic_numbers(df, condition, multi_iso_condition, l):

    genes = 0
    mrna = 0
    raw = 0
    total = 0
    shared = 0

    for j, row in df.iterrows():

        if int(row['mRNA_nb']) > 1:
            mrna += int(row['mRNA_nb'])

        cds_row = eval(row['CDS_locations_corrected'])
        previous_cds = (0, 0)
        for cds, cds_content in cds_row.items():
            raw += 1
            if previous_cds[1] != cds[0]:
                total += 1
                if cds_content['Sharing_status'] == True:
                    shared += 1
                
            previous_cds = cds
        genes += 1

    l.append(
        {
        'Total/Multi-iso' : multi_iso_condition,
        'Condition' : condition,
        'Genes' : genes,
        'mRNA' : mrna,
        'Unique exons nb' : total,
        'Shared exons nb' : shared,
        }
    )


m = multiprocessing.Manager()
l = m.list()
# for case, condition, multi_iso_condition in zip([full_human_genes, full_human_genes.loc[full_human_genes["mRNA_nb"] > 1]], ['Total', 'Total'], ['Total', 'Multi-iso']):
#     l.append(basic_numbers(case, condition, multi_iso_condition))

parmap.starmap(
    basic_numbers, 
    list(zip(
        [full_human_genes, full_human_genes.loc[full_human_genes["STATUS"] == 'DISEASE'], full_human_genes.loc[full_human_genes["STATUS"] == 'HEALTHY'], full_human_genes.loc[full_human_genes["mRNA_nb"] > 1], full_human_genes.loc[(full_human_genes["STATUS"] == 'DISEASE') & (full_human_genes["mRNA_nb"] > 1)], full_human_genes.loc[(full_human_genes["STATUS"] == 'HEALTHY') & (full_human_genes["mRNA_nb"] > 1)]], 
        ['Total', 'Disease', 'Healthy', 'Total', 'Disease', 'Healthy',], 
        ['Total', 'Total', 'Total', 'Multi-iso', 'Multi-iso', 'Multi-iso'])
        ), 
    l, 
    pm_pbar=True
)
df = pd.DataFrame(list(l))
df['Total/Multi-iso'] = pd.Categorical(df['Total/Multi-iso'], 
                        categories=["Total", 'Multi-iso'],
                    )
df['Condition'] = pd.Categorical(df['Condition'], 
                    categories=["Total", 'Disease', 'Healthy'],
                    )

df.sort_values(by=['Total/Multi-iso', 'Condition']).reset_index(drop=True)

#TODO : add to excel sheet
```

```python
def count_cds(row):
    d = {
        'Variable_region':
            {
                'Constitutive': 0,
                'Alternative': 0,
            },
        'Unique':
            {
                'Constitutive': 0,
                'Alternative': 0,
            }
    }

    row = eval(row)

    for cds, cds_content in row.items():
        if eval(cds_content['Ratio']) < 1:
            d[cds_content['CDS_representation']]['Alternative'] += 1
        else:
            d[cds_content['CDS_representation']]['Constitutive'] += 1

    const = d['Variable_region']['Constitutive'] + d['Unique']['Constitutive']
    alt = d['Variable_region']['Alternative'] + d['Unique']['Alternative']

    Unalterable = d['Unique']['Constitutive'] + d['Unique']['Alternative']
    Modifiable = d['Variable_region']['Constitutive'] + d['Variable_region']['Alternative']

    Unalterable_constitutive = d['Unique']['Constitutive']
    Unalterable_alternative = d['Unique']['Alternative']

    Modifiable_constitutive = d['Variable_region']['Constitutive']
    Modifiable_alternative = d['Variable_region']['Alternative']

    return pd.Series([const, alt, const + alt, Unalterable, Modifiable, Unalterable_constitutive, Unalterable_alternative, Modifiable_constitutive, Modifiable_alternative])


def return_tp53_like_genes(refseq, disease_genes):
    refseq = refseq.loc[(refseq['mRNA_nb'] > 1)]
    refseq[[
        'Count_CDS_constitutive', 'Count_CDS_alternative', 'Count_CDS', 'Count_CDS_Unalterable', 'Count_CDS_Variable', 'Count_CDS_Unalterable_Constitutive', 
        'Count_CDS_Unalterable_Alternative', 'Count_CDS_Modifiable_Constitutive', 'Count_CDS_Modifiable_Alternative'
        ]] = refseq['CDS_locations_corrected'].apply(lambda r: count_cds(r))
    return refseq

# df_transformed = pd.read_csv(project + 'data/2_processed/DF_transformed.csv.gz', compression='gzip', sep='\t')
df_transformed = refseq.rename(
    {
        'Gene' : 'CCRS_Gene',
        'Count_CDS_alternative' : 'RefSeq_Count_CDS_alternative',
        'Count_CDS_constitutive' : 'RefSeq_Count_CDS_constitutive',
        'End' : 'RefSeq_End',
        'GeneID' : 'RefSeq_GeneID',
        'HGNC' : 'RefSeq_HGNC',
        'Length' : 'RefSeq_Length',
        'Ratio' : 'RefSeq_Ratio',
        'Sharing_status' : 'RefSeq_Sharing_status',
        'Sources' : 'RefSeq_Sources',
        'Start' : 'RefSeq_Start',
        'mRNA_IDS' : 'RefSeq_mRNA_IDS',
        'mRNA_nb' : 'RefSeq_mRNA_nb',
        'ranges' : 'RefSeq_ranges',
    },
    axis=1,
)


# Read complete RefSeq lite file
complete_data = full_human_genes.copy()
#     complete_data = complete_data.dropna(subset=['HGNC'])
#     complete_data['HGNC'] = complete_data['HGNC'].astype(np.int64)
# complete_data_disease = complete_data.loc[complete_data['HGNC'].isin(disease_genes)]
# complete_data_healthy = complete_data.loc[~complete_data['HGNC'].isin(disease_genes)]
complete_data.loc[complete_data['Name'].isin(disease_genes), 'STATUS'] = 'DISEASE'
complete_data.loc[~complete_data['Name'].isin(disease_genes), 'STATUS'] = 'HEALTHY'
complete_data[['Count_CDS_constitutive', 'Count_CDS_alternative', 'Count_CDS', 'Count_CDS_Unalterable', 'Count_CDS_Variable', 'Count_CDS_Unalterable_Constitutive',  'Count_CDS_Unalterable_Alternative', 'Count_CDS_Modifiable_Constitutive', 'Count_CDS_Modifiable_Alternative']] = complete_data['CDS_locations_corrected'].apply(lambda r: count_cds(r))
```
